```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
Invest2_scaled = scaler.fit_transform(Invest2) #activation function of neural layer is   relu requires inputs from -1 to 1 

Return = Return.values.reshape(-1,1) # in order to allow for the 
Return_scaled = scaler.fit_transform(Return)
```

```python
from sklearn.model_selection import train_test_split 
Invest2_train, Invest2_test, Return_train, Return_test = train_test_split(Invest2_scaled, Return_scaled, test_size = 0.25, random_state = 10)


#Invest2_valtrain, Invest2_valtrain, Return_valtrain, Return_valtest = train_test_split(Invest2_scaled, Return_scaled, test_size = 0.20, random_state = 10)
```

```python
import tensorflow.keras 
from keras.models import Sequential 
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
model = Sequential()

model.add(Dense(20, input_dim = 10, activation = 'relu'))
model.add(Dense(22, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='linear'))# outputs a single value value and compatible with regression
model.summary()
```

```python
model.compile(optimizer = "adam", loss = 'mean_squared_error') 
# adam optimizer because it is shown to have the best efficiency when it comes to training a model compared to other optimizers
```
